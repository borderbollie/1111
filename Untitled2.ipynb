```python
!pip install finlab
from finlab import data
df=data.get('price:')


df[['0050','0056','2303','2330','2317']]
stocks=['0050','0056','2303','2330','2454','2881','2542','00662']

df1=df[stocks].dropna()


df2=df1['2020-01-01':'2023-07-10']


import matplotlib.pyplot as plt

((df2.pct_change()+1).cumprod()).plot(figsize=(10, 7))
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
