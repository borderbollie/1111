```python
import numpy as np
from  matplotlib import pyplot as plt

x = np.arange(1,20)
y = x*x+x

plt.title("matplotlib demo")
plt.xlabel("x軸")
plt.ylabel("y軸")
plt.plot(x)
plt.plot(y)
```
